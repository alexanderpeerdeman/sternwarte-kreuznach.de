```python
import yaml
import os
import frontmatter
from io import BytesIO
```

```python
root = "../content/vhs"
vhs_files = []
for path, subdirs, files in os.walk(root):
    for name in files:
        if name.endswith(".md") and name not in ["_index.md", "archiv.md"]:
            vhs_files.append(os.path.join(path, name))
print(len(vhs_files))
```

```python
for file in vhs_files:
    post = frontmatter.load(file)
    
    try:
        speakers = post["talk_speakers"]
    except KeyError:
        continue
    
    new_format = dict()
    for index, speaker in enumerate(speakers):
        new_format["speaker{}".format(index+1)] = {
            "name": speaker
        }

    post["talk_speakers"] = new_format

    with open(file, "wb") as f:
        print(frontmatter.dump(post, f))
```
